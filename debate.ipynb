```python
from dotenv import load_dotenv
import os
import requests
import re

# Load environment variables from .env file
load_dotenv()

# Retrieve the API keys and user IDs from environment variables
api_key_elon = os.getenv("API_KEY_ELON")
user_id_elon = os.getenv("USER_ID_ELON")
api_key_yann = os.getenv("API_KEY_YANN")
user_id_yann = os.getenv("USER_ID_YANN")

conversation=[
    {
        "speaker": "Yann",
        "text": "Open research publications are crucial for advancing AI. Sharing knowledge accelerates progress and ensures ethical development of technology. Look at the breakthroughs we've achieved through collaborative efforts like arXiv and open-source projects."
    },
    {
        "speaker": "Elon",
        "text": "While openness has merits, some research must remain private. Proprietary tech gives companies a competitive edge and drives innovation. Tesla's self-driving technology, for instance, benefits from keeping certain advancements under wraps."
    },
    {
        "speaker": "Yann",
        "text": "But secrecy can lead to dangerous monopolies in AI. Open collaboration allows for peer review and prevents potentially harmful applications. Remember the Cambridge Analytica scandal? That's what happens when algorithms develop in the dark."
    },
    {
        "speaker": "Elon",
        "text": "Peer review has its place, but business requires some level of secrecy. Without it, there's little incentive for companies to invest in R&D. SpaceX wouldn't exist if we had to publish every rocket design."
    },
    {
        "speaker": "Yann",
        "text": "The risks of secretive AI development outweigh potential profits. We need transparency to address bias, safety, and ethical concerns. The AI community has made significant strides in fairness and accountability through open discourse."
    },
    {
        "speaker": "Elon",
        "text": "Balance is key. We can have both: open research for foundational work, and private projects to turn that research into viable products. It's not black and white; there's room for a hybrid approach."
    },
    {
        "speaker": "Yann",
        "text": "A hybrid approach sounds good in theory, but where do we draw the line? Keeping core AI research open is vital for democratic development of this powerful technology. We can't let it be controlled by a few corporations."
    },
    {
        "speaker": "Elon",
        "text": "I agree on avoiding monopolies, but complete openness could lead to bad actors misusing the technology. Some level of responsible gatekeeping is necessary. Plus, competition between companies often leads to faster innovation."
    },
    {
        "speaker": "Yann",
        "text": "Fast innovation without proper checks can be dangerous. Open research allows for diverse perspectives and catches potential issues early. It also prevents duplication of efforts, saving time and resources globally."
    },
    {
        "speaker": "Elon",
        "text": "Valid points, but consider national security implications. Some AI research, if made public, could pose risks in the wrong hands. There needs to be a mechanism to protect sensitive advancements."
    },
    {
        "speaker": "Yann",
        "text": "National security is important, but it's often used as an excuse for excessive secrecy. Most AI research doesn't fall into this category. Openness actually improves security by allowing many eyes to spot vulnerabilities."
    },
    {
        "speaker": "Elon",
        "text": "Perhaps we can agree on a framework: open research for general principles and algorithms, with specific implementations and applications allowed to be proprietary. This could balance progress, security, and business interests."
    }
]

# Headers for each speaker
headers_elon = {
    "accept": "text/event-stream",
    "content-type": "application/json",
    "X-User-Id": user_id_elon,
    "Authorization": f"Bearer {api_key_elon}"
}

headers_yann = {
    "accept": "text/event-stream",
    "content-type": "application/json",
    "X-User-Id": user_id_yann,
    "Authorization": f"Bearer {api_key_yann}"
}

# Voice IDs for each speaker
elon_voice_id = "s3://voice-cloning-zero-shot/34f34028-2be3-4ee9-9d62-b7f8e1d03019/elon/manifest.json"
yann_voice_id = "s3://voice-cloning-zero-shot/6b30e2c1-40c2-4409-8b9b-098c5c1d34e4/yann-lecun/manifest.json"

# Function to generate audio for each piece of text
def generate_audio(text, headers, voice_id):
    url = "https://api.play.ht/api/v2/tts"
    payload = {
        "text": text,
        "voice": voice_id,
        "output_format": "mp3",
        "voice_engine": "PlayHT2.0"
    }
    response = requests.post(url, json=payload, headers=headers)
    return response.text

# Function to extract the final audio URL from the response text
def extract_audio_url(response_text):
    match = re.search(r'"url":"(https://[^"]+)"', response_text)
    if match:
        return match.group(1)
    return None

# List to store the results
results = []

# Generate audio for each piece of text
for entry in conversation:
    speaker = entry["speaker"]
    text = entry["text"]
    if speaker == "Elon":
        headers = headers_elon
        voice_id = elon_voice_id
    else:
        headers = headers_yann
        voice_id = yann_voice_id
    
    response_text = generate_audio(text, headers, voice_id)
    audio_url = extract_audio_url(response_text)
    if audio_url:
        results.append({"speaker": speaker, "text": text, "audio_url": audio_url})
        print(f"Audio URL for {speaker}: {audio_url}")

# Print all results
print("All Results:", results)
```

```python
#I created an AI pipeline with which you can create debate on a any topic and gets narrated by your voice clone. Here you can listen to an debate between kanye west and elon musk on the recent crowdstrike IT outages: 
import requests
from pydub import AudioSegment
import os

# Directory to save downloaded MP3 files
download_dir = "downloaded_mp3s"
os.makedirs(download_dir, exist_ok=True)

# Download MP3 files
for i, result in enumerate(results):
    audio_url = result["audio_url"]
    response = requests.get(audio_url)
    file_path = os.path.join(download_dir, f"audio_{i}.mp3")
    with open(file_path, "wb") as file:
        file.write(response.content)
    result["file_path"] = file_path

# Create separate audio tracks for each speaker
elon_audio = AudioSegment.silent(duration=0)
yann_audio = AudioSegment.silent(duration=0)

for result in results:
    audio = AudioSegment.from_mp3(result["file_path"])
    if result["speaker"] == "Elon":
        elon_audio += audio
        yann_audio += AudioSegment.silent(duration=len(audio))
    else:
        yann_audio += audio
        elon_audio += AudioSegment.silent(duration=len(audio))

# Export the individual audio tracks to separate MP3 files
elon_audio.export("elon_audio_final_final.mp3", format="mp3")
yann_audio.export("yann_audio_final_final.mp3", format="mp3")

print("Elon audio saved as 'elon_audio_final_final.mp3'")
print("Yann audio saved as 'yann_audio_final_final.mp3'")
```

```python
#put all generated audios together
import requests
from pydub import AudioSegment
import os

# Directory to save downloaded MP3 files
download_dir = "downloaded_mp3s"
os.makedirs(download_dir, exist_ok=True)

# Download MP3 files
for i, result in enumerate(results):
    audio_url = result["audio_url"]
    response = requests.get(audio_url)
    file_path = os.path.join(download_dir, f"audio_{i}.mp3")
    with open(file_path, "wb") as file:
        file.write(response.content)
    result["file_path"] = file_path

# Combine MP3 files
combined = AudioSegment.empty()
for result in results:
    audio = AudioSegment.from_mp3(result["file_path"])
    combined += audio

# Export the combined audio to a single MP3 file
combined.export("combined_audio_isolated_crowdstrike.mp3", format="mp3")

print("Combined audio saved as 'combined_audio_isolated.mp3'")
```
