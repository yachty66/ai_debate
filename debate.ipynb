```python
from dotenv import load_dotenv
import os
import requests
import re

# Load environment variables from .env file
load_dotenv()

# Retrieve the API keys and user IDs from environment variables
api_key_speaker_1 = os.getenv("API_KEY_MAX")
user_id_speaker_1 = os.getenv("USER_ID_MAX")
api_key_speaker_2 = os.getenv("API_KEY_WOBBERT")
user_id_speaker_2 = os.getenv("USER_ID_WOBBERT")

conversation=[
    {
        "speaker": "Julius Caesar",
        "text": "By Jupiter's beard, what manner of man art thou? Your head looks as though it's been attacked by a flock of mischievous birds!"
    },
    {
        "speaker": "Mark Zuckerberg",
        "text": "Uh, hello. I'm Mark Zuckerberg, founder of Facebook. It's an honor to meet you, Julius Caesar."
    },
    {
        "speaker": "Julius Caesar",
        "text": "Facebook? Is that some new province I've yet to conquer? And what's that gaudy trinket hanging 'round your neck? It looks like something Cleopatra would scoff at!"
    },
    {
        "speaker": "Mark Zuckerberg",
        "text": "This? It's just a gold chain. It's... fashion."
    },
    {
        "speaker": "Julius Caesar",
        "text": "Fashion? Bah! In Rome, we adorn ourselves with laurel wreaths, symbols of victory and honor. Your curly mop and shiny neckpiece make you look like a jester in Neptune's court!"
    },
    {
        "speaker": "Mark Zuckerberg",
        "text": "I assure you, I'm not a jester. I run one of the largest tech companies in the world."
    },
    {
        "speaker": "Julius Caesar",
        "text": "Tech? Company? Your strange words mean nothing to me, curly one. Tell me, can your 'tech company' conquer Gaul? Can it cross the Rubicon?"
    },
    {
        "speaker": "Mark Zuckerberg",
        "text": "Well, no, but we connect billions of people across the globe. We've revolutionized communication."
    },
    {
        "speaker": "Julius Caesar",
        "text": "Billions? Preposterous! There aren't even a million people in the known world. Your curls must be affecting your reason, young Mark."
    },
    {
        "speaker": "Mark Zuckerberg",
        "text": "I'm telling the truth. The world is much bigger now, and we've advanced a lot since your time."
    },
    {
        "speaker": "Julius Caesar",
        "text": "Advanced? Ha! You, with your soft hands and strange attire? I've seen more intimidating figures carved on the side of a Greek urn! At least lose the chain, boy. It makes you look like a pompous patrician who's never seen a day of battle."
    },
    {
        "speaker": "Mark Zuckerberg",
        "text": "I... I'll consider it. Maybe I should look into laurel wreaths instead."
    },
    {
        "speaker": "Julius Caesar",
        "text": "Now you're talking sense, despite your ridiculous hair. Perhaps there's hope for you yet, young Zuckerberg. Now, tell me more about this 'Facebook' of yours. Can it be used to spread news of my greatness across your billions?"
    }
]

# Headers for each speaker
headers_speaker_1 = {
    "accept": "text/event-stream",
    "content-type": "application/json",
    "X-User-Id": user_id_speaker_1,
    "Authorization": f"Bearer {api_key_speaker_1}"
}

headers_speaker_2 = {
    "accept": "text/event-stream",
    "content-type": "application/json",
    "X-User-Id": user_id_speaker_2,
    "Authorization": f"Bearer {api_key_speaker_2}"
}

# Voice IDs for each speaker
speaker_1_voice_id = "s3://voice-cloning-zero-shot/34f34028-2be3-4ee9-9d62-b7f8e1d03019/elon/manifest.json"
speaker_2_voice_id = "s3://voice-cloning-zero-shot/6b30e2c1-40c2-4409-8b9b-098c5c1d34e4/yann-lecun/manifest.json"

# Function to generate audio for each piece of text
def generate_audio(text, headers, voice_id):
    url = "https://api.play.ht/api/v2/tts"
    payload = {
        "text": text,
        "voice": voice_id,
        "output_format": "mp3",
        "voice_engine": "PlayHT2.0"
    }
    response = requests.post(url, json=payload, headers=headers)
    return response.text

# Function to extract the final audio URL from the response text
def extract_audio_url(response_text):
    match = re.search(r'"url":"(https://[^"]+)"', response_text)
    if match:
        return match.group(1)
    return None

# List to store the results
results = []

# Generate audio for each piece of text
for entry in conversation:
    speaker = entry["speaker"]
    text = entry["text"]
    if speaker == "Elon":
        headers = headers_speaker_1
        voice_id = speaker_1_voice_id
    else:
        headers = headers_speaker_2
        voice_id = speaker_2_voice_id
    
    response_text = generate_audio(text, headers, voice_id)
    audio_url = extract_audio_url(response_text)
    if audio_url:
        results.append({"speaker": speaker, "text": text, "audio_url": audio_url})
        print(f"Audio URL for {speaker}: {audio_url}")

# Print all results
print("All Results:", results)
```

```python
#I created an AI pipeline with which you can create debate on a any topic and gets narrated by your voice clone. Here you can listen to an debate between kanye west and elon musk on the recent crowdstrike IT outages: 
import requests
from pydub import AudioSegment
import os

# Directory to save downloaded MP3 files
download_dir = "downloaded_mp3s"
os.makedirs(download_dir, exist_ok=True)

# Download MP3 files
for i, result in enumerate(results):
    audio_url = result["audio_url"]
    response = requests.get(audio_url)
    file_path = os.path.join(download_dir, f"audio_{i}.mp3")
    with open(file_path, "wb") as file:
        file.write(response.content)
    result["file_path"] = file_path

# Create separate audio tracks for each speaker
elon_audio = AudioSegment.silent(duration=0)
yann_audio = AudioSegment.silent(duration=0)

for result in results:
    audio = AudioSegment.from_mp3(result["file_path"])
    if result["speaker"] == "Elon":
        elon_audio += audio
        yann_audio += AudioSegment.silent(duration=len(audio))
    else:
        yann_audio += audio
        elon_audio += AudioSegment.silent(duration=len(audio))

# Export the individual audio tracks to separate MP3 files
elon_audio.export("elon_audio_final_final.mp3", format="mp3")
yann_audio.export("yann_audio_final_final.mp3", format="mp3")

print("Elon audio saved as 'elon_audio_final_final.mp3'")
print("Yann audio saved as 'yann_audio_final_final.mp3'")
```

```python
#put all generated audios together
import requests
from pydub import AudioSegment
import os

# Directory to save downloaded MP3 files
download_dir = "downloaded_mp3s"
os.makedirs(download_dir, exist_ok=True)

# Download MP3 files
for i, result in enumerate(results):
    audio_url = result["audio_url"]
    response = requests.get(audio_url)
    file_path = os.path.join(download_dir, f"audio_{i}.mp3")
    with open(file_path, "wb") as file:
        file.write(response.content)
    result["file_path"] = file_path

# Combine MP3 files
combined = AudioSegment.empty()
for result in results:
    audio = AudioSegment.from_mp3(result["file_path"])
    combined += audio

# Export the combined audio to a single MP3 file
combined.export("combined_audio_isolated_crowdstrike.mp3", format="mp3")

print("Combined audio saved as 'combined_audio_isolated.mp3'")
```
